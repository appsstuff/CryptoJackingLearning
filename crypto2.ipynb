```python
import sklearn
print(sklearn.__version__)  # Print the installed scikit-learn version

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
# Choose your preferred model (e.g., Random Forest, SVM, Gradient Boosting)
from sklearn.ensemble import RandomForestClassifier  # Example model
import numpy as np  # Import NumPy for missing value handling (optional)
import matplotlib.pyplot as plt  # Optional for data exploration
```

```python
# 1. Load and Preprocess Data (with Optional Exploration)
def load_data(filepath):
  """Loads data from a CSV file, performs basic preprocessing, and offers optional exploration."""
  data = pd.read_csv(filepath)
  data["target"]=0
  # Handle missing values (replace with your specific logic)
  # Example: fill missing values with mean/median or remove rows with many missing values
  data = data.fillna(method='ffill')  # Fill missing values with previous value (example)
  # data = data.dropna(axis=0, thresh=0.8)  # Drop rows with more than 20% missing values (example)

  # Select relevant features (replace with your chosen features)
  features = ['CPUCore#1', 'CPUCore#2', 'CPUCore#3', 'CPUCore#4', 'CPUTotal', 'Memory', 'UsedMemory','target']
  X = data[features]
  y = data['Status']

  # Optional data exploration (uncomment if desired)
  # Use box plots or other visualizations to explore feature distributions for each class
  # grouped_data = data.groupby("Status")
  # grouped_data.boxplot(subplots=True, layout=(4, 4), figsize=(12, 12))
  # plt.suptitle("Distribution of Features by Status")
  # plt.show()

  return X, y
```

```python
# 2. Train-Test Split with Class Weighting
def train_test_split_with_weights(X, y):
  """Splits data into training and testing sets, calculates class weights."""
  try:
      class_weights = compute_class_weight(class_weight='balanced', classes=np.unique(y), y=y)
      X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.40, random_state=0)
      return X_train, X_test, y_train, y_test, class_weights
  except Exception as e:
        raise ValueError("Error computing class weights: {}".format(e))

```

```python
# 3. Model Training
def train_model(X_train, y_train):
  """Trains a model with class weighting."""
  model = RandomForestClassifier(n_estimators=250 , random_state=0)
  model.fit(X_train, y_train)
  return model
```

```python


# 4. Evaluation and Prediction
def evaluate_and_predict(model, X_test, y_test):
  """Evaluates model performance and makes predictions on unseen data."""
  from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

  # Make predictions
  y_pred = model.predict(X_test)

  # Calculate metrics
  accuracy = accuracy_score(y_test, y_pred)
  precision = precision_score(y_test, y_pred)
  recall = recall_score(y_test, y_pred)
  f1 = f1_score(y_test, y_pred)

  # Print evaluation metrics (including accuracy)
  print(f"Accuracy: {accuracy:.4f}")
  print(f"Precision: {precision:.4f}")
  print(f"Recall: {recall:.4f}")
  print(f"F1-Score: {f1:.4f}")

  # Return predictions (optional)
  return y_pred
```

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=0)

XClass, YClass = make_classification(n_samples=10000, n_features=2, n_redundant=0,
	n_clusters_per_class=1, weights=[0.99], flip_y=0, random_state=1)
oversample = RandomOverSampler(sampling_strategy='minority')
Xmodel, Ymodel = oversample.fit_resample(XClass, YClass)
# 5. Driver Code (Example Usage)

X_train, X_test, Y_train, Y_test = train_test_split(Xmodel, Ymodel, test_size=0.24, random_state=1)
model = train_model(X_train, Y_train)
prediction = model.predict(X_test)
prediction = evaluate_and_predict(model, X_test, Y_test)
```

```python
def plotPrediction(predict,label , score, target):
  label = label + " , Score: " + str(score )
  plt.scatter(Y_test[target].values, Y_test, color = 'red')
  plt.scatter(Y_test[target].values, predict, color = 'blue')
  plt.title(label)
  plt.xlabel('Status')
  plt.ylabel('Prediction')
  plt.legend
  plt.show()
```
